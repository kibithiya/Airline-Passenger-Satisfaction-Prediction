```python
df['Arrival Delay in Minutes']=df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].mean)
```

```python
df.isna().sum()
```

```python
categorical_columns=['Gender','Customer Type','Type of Travel','Class','satisfaction']
for column in categorical_columns:
    print(f'Unique values in {column} are :{df[column].unique()}')
    
```

```python
def zeroorone(x):
    if x=='neutral or dissatisfied':
        return 0
    else:
        return 1
    
    
```

```python
df['satisfaction']=df['satisfaction'].apply(zeroorone)
```

```python
x=pd.get_dummies(df,drop_first=True)
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
x_train.size
```

```python
y_train.size
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=40,max_depth=1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python

from sklearn.metrics import  accuracy_score,confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
c=confusion_matrix(y_test,y_pred) #yaxis actual value in plot and x axis predcted value
```

```python
sns.heatmap(c,annot=True,fmt=".1f")
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```
